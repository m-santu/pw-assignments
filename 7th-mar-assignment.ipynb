> `Mean` used to measure the average of the dataset

> `Median` used to measure the middle value of dataset

> `Mode` used to measure the most frequent data in the dataset

## 

```python
import pandas as pd
```

```python
dataset = pd.Series([178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5])

mean = dataset.mean()
median = dataset.median()
mode = dataset.mode().tolist()

print(f"mean : {mean}\nmedian : {median}\nmode : {mode}")
```

## 

formula for calculation sample mean :
> sum of all values in the sample dataset devide by it's size. (x1 + x2 + x3 + x4 + ...) / n

```python
## suppose we have a dataset [1, 2, 3, 4, 5]. and wants to calculate it's mean
dataset = [1, 2, 3, 4, 5]

mean = sum(dataset) / len( dataset )
print(f'mean : {mean}')
```

## 

in Measures of Central Tendency, the mean is most effected by ouliers, median or mode doesn't effected that much. In the presence of outliers the mean value tends toword the outlier.

In dataset [1, 2, 3, 3, 4, 3, 1000] the presence of outlier 1000 shift a lot. the mean of the dataset with outliar is 145, and without outliar 2.6666

Outliears even effect the dispersion a lot. Outliers can increase the variability of the data, resulting in a larger variance or standard deviation.

```python
## standard deviation without outliers

d1 = [1, 2, 3, 3, 4, 3]
np.std(d1)
```

```python
## standard deviation with outliers

d2 = [1, 2, 3, 3, 4, 3, 1000]
np.std(d2)
```

the gap between 1st and 2nd standard deviation is huge, just by adding one outliar
