## 

`MongoDB` : MongoDB is a type of non-relational database

non-relation database is type of database where we store unstructure data

> Choose MongoDB ocer SQL in case :
>- we have unstructure data
>- easy to scale
>- read and write is faster

---
<br></br>
## 

MongoDB Features :
- `Document Oriented` : MongoDB stores document oriented data, for example JSON
- `Flexible Schema` : MongoDB provides very flexible schema. means every document can have different fields
- `Authentication` : MongoDB provides a well authentication mechanisms to access the database. Bydefault is the Salted Challenge Response Authentication Mechanism (SCRAM)
- `Indexing` : in sort indexing is a special kind of data structure that give the ability to quick search and retrieve information within our collection

## 

```python
def document( name, class_name, roll ) :
    return { "name" : name, "class" : class_name, "roll" : roll }
```

```python
# inserting one record to the collection
collection.insert_one( document( 'rohan', 'V', 1 ) )
```

```python
data = [
    document('akash', 'X', 3),
    document('rohit', 'X', 2),
    document('Sohoam', 'XI', 32)
]
```

```python
#insert many record to the collection
collection.insert_many(data)
```

<br></br>
Fetch One document from collection 'student'

```python
fetch_one = collection.find_one()
print(fetch_one)
```

<br></br>
Fetch all documents from collection 'student'

```python
fetch_all = collection.find()
```

```python
for doc in fetch_all :
    print(doc)
```

## 

We just have to pass a query dictionary {query-key : query-value} as argument to find() function to query the MongoDB database

```python
#query from 'student' collection
query = { "class" : "X" }

fetch_query = collection.find(query)

for doc in fetch_query :
    print(doc)
```

---
<br></br>
## 

> The sort() method specifies the order in which the query returns the matching documents from the given collection




```python
'''sorting result by 'roll' in descending order'''
fetch_sorted_data = collection.find().sort('roll', pymongo.DESCENDING)
```

```python
for doc in fetch_sorted_data :
    print(doc)
```
