## 

`API` : API stands for 'Application Programming Interface'. It is a mechanisms that enable two software components to communicate with each other using a set of definitions and protocols.

Google Maps is a real world example of API. where Google provide an API of Google Map for developer who will integrate with their own application. This API provides developers with access to a lots of map-based services, such as displaying maps, retrieving information about locations etc. 'Ola', 'Uber' use the API to provide users with real-time information about their driver's location and estimated arrival time.

## 

> `Advantages` : 
>- Once an API was created and published, then many other application can use it without creating softwere for it's own
>- developers can avoid the need to build complex functionality from scratch, which can save time, reduce costs, and improve overall efficiency.
>- Any kind of programing language support API
>- APIs provide a standardized way for different applications and services to communicate with each other

> `Disadvantages` :
>- using API make our programme dependent on that APIs performance. It can create issues like downtime or service disruptions.
>- When useing API developer have to dependent on the provider of that API for maintance
>- If APIs got update without supporting previous version, then developer have to structure his code again for support the new changes
>- althow some API is free but most of the API are costly

## 

> `Web APIs` : Web APIs are used to enable communication and data exchange between different web-based applications and services. It designed to be accessed over HTTP (Hypertext Transfer Protocol) or other web protocols

> The main difference is Web APIs are only designed to access over web-protocall (such as HTTP). while APIs can communicate with all protocols. Web APIs are a type of APIs.
> Web APIs used mainly JSON or XML to communicate. but APIs may or may not be communication with JSON or XML

## 

> `REST` : it is a stateless architecture used to communicate with client and server over HTTP. It uses web function like GET, PUT, POST, and DELETE.

> `SOAP` : This architecture used to exchanging structured information between distributed systems over the internet over XML. SOAP messages always have a header and a body written mostly in XML

Shortcomings of SOAP :
- SOAP may communicate with different application but it's very complex to write
- SOAP is havy compare to REST architecture
- Not suited for all web based application
- XML Injection is a Security Threat in SOAP
- Complex and Dificult to scale the application written in SOAP

---
<br></br>
## 

> Difference Between REST and SOAP :
>- REST relay on HTTP while SOAP relay on XML
>- REST is light weight and much simpler than SOAP. SOAP is very complex and heavy
>- We can scale REST based application easily compare SOAP application based
>- Most web based application support REST architecture over SOAP
>- As Message formate REST use JSON while SOAP use XML that must include a header and body
>- REST architecture is faster than SOAP
