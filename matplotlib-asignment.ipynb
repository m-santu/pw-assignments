## 

`scatter plot` : scatter plots are useful for visualizing the relationship between two variables (on different axis) using points in the graph and identifying any patterns or trends in the data.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x, y, c='r')
plt.title('This is ans of Q2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

## 

`subplot()` function mostly use to create a multiple plot within a same figure.

```python
x = np.array([0, 1, 2, 3, 4, 5]) # x is same for all 4 lines
y1 = np.array([0, 100, 200, 300, 400, 500]) # y values for 1st line
y2 = np.array([50, 20, 40, 20, 60, 70]) # y values for 2nd line
y3 = np.array([10, 20, 30, 40, 50, 60]) # y values for 3rd line
y4 = np.array([200, 350, 250, 550, 450, 150]) # y values for 4th line

plt.figure(figsize=(9, 6))

ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4)

# plotting
ax1.plot(x, y1, 'r')
ax2.plot(x, y2, 'purple')
ax3.plot(x, y3, 'r')
ax4.plot(x, y4, 'purple')

# title
ax1.set_title('Plot For 1st Line')
ax3.set_title('Plot For 3rd Line')
ax2.set_title('Plot For 2nd Line')
ax4.set_title('Plot For 4th Line')

# xlabel
ax1.set_xlabel('X Values')
ax2.set_xlabel('X Values')
ax3.set_xlabel('X Values')
ax4.set_xlabel('X Values')

# ylabel
ax1.set_ylabel('Y Values')
ax2.set_ylabel('Y Values')
ax3.set_ylabel('Y Values')
ax4.set_ylabel('Y Values')



plt.tight_layout(pad=3)
plt.show()
```

## 

`bar plot` : this type of plot represent the relation between catagoric and numeric variable. every bar is rectangular and have it's own length or height that shows the value of the catagory

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
''' 1. vertical bar plot '''
plt.bar(company, profit, color='purple')
plt.title('This is a bar plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
''' 2. horizontal bar plot '''
plt.barh(company, profit, color='purple')
plt.title('This is a horizontal bar plot')
plt.ylabel('Company')
plt.xlabel('Profit')
plt.grid()
plt.show()
```

## 

> `box plot` : A box plot is a type of data visualization that is used to represent the distribution of a dataset

> box plot use to compare the distributions of datasets or for identifying outliers in dataset.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2])
plt.title('Box Chart')
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.show()
```
