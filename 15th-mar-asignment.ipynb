## 

### Supervised Learning
> It is a type of Machine Learning in which machines are trained using well "labelled" training data. "labelled" means some input data are mapped with it's correct ouput. Here models are supervised using training dataset

#### List of Example :
    - Test Recognization
    - Cancer Detection
    - Spam Detection
    - Image Classification
    - Voice Recognization

---
<br></br>
## 

### Unsupervised Learning
> Here models trained with "unlabelled" data set. It tries to find patterns, structure, or relationships in the data on its own.

#### List of Example :
    - Grouping News Based on Topic
    - Finding unusal pattern in Medical Sector
    - Based on previous stock history identify similar stocks

## 

> 1. Training Set : This is a subset of entire dataset that used to train our model so that our model can predict the outcom over unseen dataset
> 2. Validation Set : This is a subset of Training Dataset to evalute our model, so that we can improve our model acuracy.
> 3. Test Set : This type of dataset is unseen for the model and used to test our model.
