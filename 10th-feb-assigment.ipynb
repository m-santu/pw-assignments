## 

|> open() used to open a file


|> different modes in python :
- `'r'` :      open for reading
- `'w'` :      open for writing, but first remove entire data and then write from the start
- `'x'` :      create a new file and open it for writing, if exist raise an error
- `'a'` :      open for writing, but append to the end of the file if it exists
- `'b'` :      use for binary file. use combine with r/w/a/x
- `'t'` :      text mode (default). use combine with r/w/a/x
- `'+'` :      open for both reading and writing. for example 'r+' mode
- `'U'` :      universal newline mode ( althow it's deprecated )

## 

`read()` : used when we want entire content of a file as a string.

```python
with open('data.txt', 'r') as f :
    data = f.read() # read entire file
    print(data)
```

<br></br>
`readline()` : used to read only a line at a time

```python
with open('data.txt', 'r') as f :
    print( f.readline() )# 1st line
    print( f.readline() )# read 2nd line
```

<br></br>
`readlines()` : read all lines and return as list of string

```python
with open('data.txt', 'r') as f :
    data = f.readlines()
    print(data)
```

## 

`write()`  :  used when we want to write a string to an open file

```python
st = 'I want to be a data scientist'

with open("file1.txt", "w") as f:
    f.write(st)
```

<br></br>
`writelines()`  :  used when we want to write a list of strings to an open file

```python
l = ['I am ankan.', ' ', 'I love python']

with open('file2.txt', 'w') as f :
    f.writelines(l)
```
