```python
import seaborn as sns
import matplotlib.pyplot as plt
```

---
<br></br>
## 

Five plots that we can plot using seaborn library :
1. Lineplot : used to show the trend or pattern of a variable over time
2. Heatmap : by different colors to represent different values we can identify patterns, similarities, differences, clusters in the distribution
3. Histogram : used to show the shape and spread of the distribution, as well as any outliers or gaps in the distribution
4. Box Plot : used to show the range, median, variability and outliers (if any) of a distribution. Used to see the skewness of a distribution
5. Scatterplot: used to visualize the relationship between two variables and to identify any patterns or trends in the data.

## 

```python
fmri = sns.load_dataset('fmri')
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, style='event', hue='region')
```

## 

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic)
```

```python
plt.figure(figsize=(8, 10))
sns.boxplot(x='pclass', y='fare', data=titanic)
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
plt.figure(figsize=(20, 8))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.show()
```
