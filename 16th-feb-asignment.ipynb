## 

`database` : a database is a structure and organize collection of data that enables efficent storage, retrival and manipulation of data

> differention of SQL and NoSQL :
>- SQL is a relation database while NoSQL is a non-relation database
>- NoSQL database highly scalable and flexible where it can support unstructure data very efficently. But SQL support only structure data
>- Schemas in SQL are fixed but in NoSQL Schemas are dynamic

## 

`DDL` : DDL stands for Data Definition Language. It is a set of SQL command for create or modify the database

<br></br>
`CREATE` : used for createing a dabase or table or view in SQL

<br></br>
`ALTER` : used for modifying existing table such as add culumn or change data type of a culumn

<br></br>
`TRUNCATE` : used to delete data from a table without deleting the table itself

<br></br>
`DROP` : used for delete existing database object like table, views etc

---
<br></br>

## 

`DML` : DML stands for 'Data Manipulation Language'. it is a set of SQL commands used to manipulate data stored in a database (like insert, update, delete)

<br></br>
`INSERT` : used to insert new data to the database table

on above example suppose we have test database and students inside it. students have 3 culumn 'name', 'roll' and 'class'

<br></br>
`UPDATE` : used to modify existing data in a database table

<br></br>
`DELETE` : specifies which rows to delete from the table.

---
<br></br>


`DQL` : DQL stands for 'Data Query Language'. used to retrieve data from database tables without modifying the data.

`SELECT` : used to retrieve data from database tables

---
<br></br>


`Primary Key` : A primary key is a column in a database table that uniquely identifies each row in that table

`Foreign Key` : A foreign key is a column in a table that refers to the primary key of another table. It is used to establish a relationship between two tables in a relational database.

---
<br></br>


<br></br>
`cursor()` : It is an database object that is used to make the connection for executing SQL queries

`execute()` : this method is used to execute a SQL query on a database using a cursor object.
