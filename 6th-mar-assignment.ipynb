## 

>### Types of Statistics :
>1. `Descriptive` : Used to organize and summarize a set of data. Suppose calculation the average weight of students in a class. Here we use Descriptive Stats
>2. `Inferential` : Used to draw conclusions or making generalizations about a population based on a sample. An Election poll is a good example of Inferential Stats. Where we ask some people whom the vote as sample. Now on based of their opinion we can draw a conclusion of entire state.

## 

`Nominal Scale` :
- works on Nominal Qualitative data type
- Order doesn't matter


<br></br>
`Ordinal Scale` :
- Works on Ordinal Qualitative data set
- Rank / Order matters a lot
- difference can't be measure

<br></br>
`Interval Scale` :
- Rank / Order matters
- doesn't have `0` (zero) starting value
- difference can be measure
- ratio cann't be measure

<br></br>
`Rational Scale` :
- Rank / Order matters
- have `0` (zero) starting value
- difference can be measure
- ration can be measure

## 

if we don't know the proper level of measurements we cant use the proper statistical method to analyze the data and it can lead us to an incorrect result and conclusions.


For example suppose a survey is conducted to investigate past 10 year temperature in Kolkata. If we analyze the data as Ratio Scale Measurement, the result may apear faulty. Because temperature doesn't have a 0 (zero) starting point like Ratio scale data. it can be -23F or -45F. So here we should use Intervale Scale Measurement, to get proper result of calculation

## 

in `Discriptive` stats we summarize and describe a dataset. where in `Inferential` stats we make conclusions about the population based on our sample analyse

Suppose we want to measure average height in a class. here we can use descriptive stats to measure the mean or the average height.

Now suppose we want to measure the average height of students in entire District. Here we should use Inferential statistics. First we analyze some sample and with those sample analyze we will try to estimate the average height of the population (students in entire district).

## 

Some common measures of central tendency are :
- Mean : calculate the average
- Median : give the middle value of the dataset
- Mode : give the most frequent value in the dataset

To measure the variability we use :
- variance : measure the spread of the values in a dataset
- standard deviation : it's a square root of variance. represent how far a data from the mean in a dataset
