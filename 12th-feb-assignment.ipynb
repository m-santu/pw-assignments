## 

`Exception`  : Exception is an python object that represent error, ocoure in runtime

Difference between Exceptions and Syntax errors :
- `Exception` can be handled (using try except block) but not `Syntax` errors
- `Syntax` error caught by interpreter before runtime but `Exception` caught in runtime

---

<br></br>
## 

|> if exception is not handled then the execution of the entire programme will stop giving an error message

```python
'''example if exception not handled'''

a = 34 / 0 # raise error

print('I wanna be a Data Scientist') # not gonna print
```

```python
'''example if exception handled'''
try :
    a = 34 / 0
except Exception as e : 
    print(e) # printing the error

#since error handled then code won't stop executing
print('I wanna be a Data Scientist')
```

## 

`try-except` used to catch and handle exceptions

```python
'''example'''
try :
    b = 3 / 0
except Exception as e :
    print(e)

print('exception handled successfully')
```

## 

```python
'''
    `try and else` :   if no exception occoure in try block then else block execute
'''

try :
    a = 4 / 2
except Exception as e :
    print(e)
else :
    print(a)
    print('no exception occoure')
```

```python
'''
    `finally` : this block will always execute either exception occoure or not
'''

try :
    a = 4 / 0
    
except Exception as e :
    print('error : ' , e)
    
finally :
    print('this will always execute')
```

```python
'''
    `raise` statement used for manually raise an exception
'''

def division(num_1, num_2) :
    if num_2 == 0 : raise ZeroDivisionError("cann't divide by 0")
    return num_1 / num_2


division(34, 0) #raise an exception
```

## 

|> Custom Exceptions made by programmer. It inherits from Exception class or it's subclass. Work exactly as built in examples

```python
'''example of custom example'''

class AgeValidationError(Exception) :
    def __init__(self, message ) :
        self.message = message

def validate_age(age) :
    if age < 0 : raise AgeValidationError('age suppose to natural number')
    elif age >= 100 : raise AgeValidationError('age should be below 100')
    return age


try :
    age = 190
    validate_age(age)

except AgeValidationError as e :
    print(e)
    
    
```

|> `custom exception` mostly used to be more specific about exception. And custom exception make easer to handle and improve error handling logic

for example in above code we try to validate the age with a custom exception `AgeValidationError`. using the custom exception we make it easer and more specific to handle that perticular kind of excpetion. And the code look clean

## 

```python
'''example of custom example'''

class AgeValidationError(Exception) :
    def __init__(self, message ) :
        self.message = message

def validate_age(age) :
    age = int(age)
    
    if age < 0 : raise AgeValidationError('age suppose to natural number')
    elif age >= 100 : raise AgeValidationError('age should be below 100')
    return age


try :
    age = validate_age( input('enter your age : ') )

except AgeValidationError as e :
    print(e)

else :
    print(f"age : {age} (valid)")
```
