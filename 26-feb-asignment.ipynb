```python
import numpy as np
list_ = [ '1' , '2' , '3' , '4' , '5' ]
array_list = np.array(object = list_)
```

---
<br></br>
### 

Yes, the difference is, `list_` points to python built-in list data structure. While `array_list` points to a numpy array

```python
''' data type of list_ variable  '''
type(list_)
```

```python
''' data type of array_list variable '''
type(array_list)
```

---
<br></br>
### 

```python
''' data type of every element in list_ '''

for i in list_ : print( type(i) )
```

```python
''' data type of every element in array_list '''

for i in array_list : print( type(i) )
```

---
<br></br>
### 

```python
array_list = np.array(object = list_, dtype = int)
```

> Yes, after changes the array_list stores only int (actually numpy.int32) type element means array of int while list_ stores the list of str

```python
''' data types of every element in array_list '''
for i in array_list : print( type(i) )
```

```python
''' data types of every element in list_ '''
for i in list_ : print( type(i) )
```
