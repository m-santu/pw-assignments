## 

`multiprocessing` : it is a module in pyhton that allow us to use` multiprocessing technique` where we can simultaneously run many process/tasks in a programme by using multiple cpu processors.

suppose we have very time consuming cpu intensive process/task which can be devided into smaller independent tasks. bydefault python doesn't allow to run simultaneously. so in python using 'multiprocessing' module we can create different 'process' where we can run those smaller tasks simultaneously. And it's improve the performance and responsiveness of our python programme

<br></br>
## 

> differences between multiprocessing and multithreading
>- multi-threading works on single cpu core while multi-processing can use multiple cpu cores
>- multi-processing give use true parallelism but multi-threading works by sharing CPU time between threads, so it's only simulates parallelism
>- for memory intensive task using threads take lot's of memory of a single cpu core while by using multi-processing we can utilize memory distribution between multiple cores

```python
from multiprocessing import Queue, Process
```

```python
def display(queue) :
    '''display data as soon any value added to the queue'''
    print('a new process started ...')
    
    n = 1
    while True :
        data = queue.get()
        if not data : break
        print(f'{n:3}. {data}')
        n+=1
    
    print('end of process')
```

```python
if __name__ == '__main__' :
    queue = Queue()
    w1 = Process( target=display, args=(queue,) )
    w1.start()
```

```python
if __name__ == '__main__' :
    while True :
        inp = input( 'enter your data (press enter without typeing to exit) : ' )
        queue.put(inp)
        if not inp : break
    w1.join()
```

## 

> `multiprocessing pool` : in python multiprocessing pool used for parallel execution of a function across multiple input values upto given number of processes

> reason for using multiprocessing pool :
>- we can specify upto how much processes we need
>- distribute tasks evenly among the worker process in the pool

## 

we use Pool() in multiprocessing module to create a pool of worker processes. it take processes as argument

```python
from multiprocessing import Pool, current_process

def test(n) :
    print(f"{current_process().name} complete it's task\n")
    
if __name__ == '__main__' :
    with Pool( processes=5 ) as pool :
        pool.map(test, range(5))
```

```python
from multiprocessing import Pool, current_process

def display(n) : 
    print(f'{current_process().name} prints number : {n}\n')

if __name__ == '__main__' :
    with Pool(processes=4) as pool :
        pool.map(display, [11, 242, 543, 767])
```
