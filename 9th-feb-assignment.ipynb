```python
class Vehicle :
    
    def __init__( self, name_of_vehicle, max_speed, average_of_vehicle ) :
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
```

<br></br>
## 

```python
class Car ( Vehicle ) :
    
    def seating_capacity( self, capacity ) :
        return self.name_of_vehicle, capacity
```

```python
bmw = Car('bmw x1', 100, 20 )
```

```python
car_name, seating_capacity = bmw.seating_capacity(4)
print(f"car name\t\t: {car_name}\nseating capacity\t: { seating_capacity }")
```

`Multiple Inheritance` : When a single child class inherits from 2 or more parent class, it's called multiple inheritance

```python
'''example of multiple inheritance'''
class School :
    def __init__(self, school_name) :
        self.school_name = school_name
    
    
class Home :
    def __init__(self, address ) :
        self.home_address = address

        
class Student( Home, School ) :
    '''this student class inherting from 2 parent class, Home and School'''
    
    def __init__(self, home_address, school_name ) :
        Home.__init__(self, home_address)
        School.__init__(self, school_name)
        
```

```python
rohon = Student( 'kolkata, west bengal', 'ABC High School')
print( rohon.school_name )
print( rohon.home_address )
```

`Getter` : in python class Getter is a method we create to get access to a private variable in class

`Setter` : in python class Setter is a method we create to modify a private variable in class

```python
''' example of a class with Getter and Setter '''

class Bank :
    def __init__(self, balance = 0) :
        self.__balance = balance if balance >= 0 else 0
        
    
    # it's a Getter method, to get current balance
    def balance(self) : 
        return self.__balance
    
    
    # Setter method to set amount while withdrawl
    def withdrawl( self, amount ) :
        if amount > self.__balance : return 'Insufficient Balance'
    
        self.__balance -= amount
        return 'Success'
    
    
    # Setter method to add balance while deposit
    def deposit(self, amount ) :
        if amount <= 0 : return 'Please Enter Proper Amount'
        
        self.__balance += amount
        return 'Success'
    
```

```python
ashim = Bank( 1000 )
```

```python
ashim.balance()
```

```python
ashim.withdrawl(500)
```

```python
ashim.balance()
```

`method overriding` : when a child class define a method same name as parent class that the child class inherit

```python
class Parent :
    def display(self) :
        print(' this is a method from Parent class ')

        
class Child( Parent ) :
    def display(self) :
        print('this is child class method')
        
```

```python
child_obj = Child()
child_obj.display()
```
