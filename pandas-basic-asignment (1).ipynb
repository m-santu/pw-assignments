```python
# creating 3 series
s1 = pd.Series(['Rohon', 'Ankan', 'Santu'], name="Name")
s2 = pd.Series([20, 22, 21], name="Age")
s3 = pd.Series(['Teacher', 'Student', 'Student'], name='Role')


# concate above series
df1 = pd.concat([s1, s2, s3], axis=1)

print(df1)
```

on above example I create 3 series and give specific name to each serieses. Then I concate along axis=1 to marge those series to create a dataframe

The created dataframe, df1 now have 3 rows and 3 columns. Like that we can create dataframe with as many serieses
