## 

- In order to catch that custom exception using `except` statement we must inherit from base Exception or it's subclass.
- Inheriting from Exception class give the custom exception the same interface same as built-in exception. means we can `raise` them manually

---

<br></br>
## 

```python
def exception_hierarchy( exception_class, tab = 0 ) :
    print('\t'*tab, exception_class.__name__)
    
    for sub_exception in exception_class.__subclasses__() :
        exception_hierarchy( sub_exception, tab + 1 )
```

> ArithmeticError class is a base class for exceptions that occur during arithmetic operations

> Some errors defined in ArithmeticError class : 
>- `FloatingPointError`
>- `OverflowError`
>- `ZeroDivisionError`
>- `DecimalException`

```python
"""
    1. `ZeroDivisionError` : Raised when attempting to divide a number by zero
"""

try :
    1234 / 0
except ZeroDivisionError as e :
    print(e)
```

> `LookupError` class is a base class for exceptions that occur when a key or index is not found in a collection

The following exceptions are defined as subclasses of LookupError, that we normaly use instade of LookupError :


```python
"""
    `IndexError` : Occoures when trying to access an index that is out of range of an Iterable or Sequence
"""

l = [1, 2, 4]

try :
    print( l[100] )
    
except IndexError as e:
    print('IndexError :', e)
```

```python
"""
    `KeyError` : Raised when trying to access a key that is not present in a dictionary
"""

d = {'name' : 'ankan'}

try :
    print( d['age'] )

except KeyError as e :
    print( 'KeyError : no such key named', e )
```

> `ImportError`  :  This type of error occurs while unable to import a module due to : 
>- The importe is in a circular dependency
>- The importe is unavailable or was not create
>- The imported module or class or function name is misspelled

```python
try :
    import Math
except ImportError as e:
    print('ImportError :',e)
```

> `ModuleNotFoundError` : this is a specific type of ImportError that occurs when Python cannot locate the module that programer trying to import.

```python
try :
    import dunno
except ModuleNotFoundError as e :
    print('ModuleNotFoundError :', e)
```
